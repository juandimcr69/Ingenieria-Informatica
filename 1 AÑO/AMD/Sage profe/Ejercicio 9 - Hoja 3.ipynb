```python
# El plano vectorial P es el espacio ortogonal al espacio generado por el vector:
```

```python
w=vector([1,-1,1]); show(w)
```

```python
# Vamos a calcular una base ortonormal {u1,v1} de P donde u1 sea proporcional al vector:
```

```python
u=vector([1,1,0]); show(u)
```

```python
# Calculamos el producto vectorial de u y w:
```

```python
v=u.cross_product(w); show(v)
```

```python
# Por tanto, {u,v} es una base ortogonal de P.
# Normalizamos los vectores para tener una base ortonormal {u1,v1} de P:
```

```python
u1=u.normalized(); v1=v.normalized(); show(u1); show(v1)
```
