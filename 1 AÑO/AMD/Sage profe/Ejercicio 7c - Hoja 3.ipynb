```python
# Sea r el espacio (recta vectorial) generado por el vector:
```

```python
u3=vector([1,2,-1]); show(u3)
```

```python
# Vamos a empezar completando u3 a una base ortogonal {u1,u2,u3} de R^3.
```

```python
# Tomamos un vector u1 (no nulo) ortogonal a u3, a ojo. Por ejemplo:
```

```python
u1=vector([1,0,1]); show(u1)
```

```python
# Ahora basta tomar como u2 el producto vectorial de u1 y u3: 
```

```python
u2=u1.cross_product(u3); show(u2)
```

```python
# Dibujamos la recta r y los vectores {u1,u2,u3}:
```

```python
var('t')
```

```python
dibujo_recta=parametric_plot(t*u3,(t,-2,2),color='red')
```

```python
plot(u1,thickness=3)+plot(u2,color='black',thickness=3)+plot(u3,color='red',thickness=3)+dibujo_recta
```

```python
# Finalmente, normalizamos los vectores para tener una base ortonormal de R^3:
```

```python
v1=u1.normalized(); v2=u2.normalized(); v3=u3.normalized(); show(v1); show(v2); show(v3)
```

```python
# Dibujamos la recta r y los vectores de esta base ortonormal:
```

```python
plot(v1,thickness=3)+plot(v2,color='black',thickness=3)+plot(v3,color='red',thickness=3)+dibujo_recta
```
