```python
# Buscamos primero una base del plano P = N((1 -1 -3)).
```

```python
# Resolviendo el sistema, obtenemos la base {u,v} siendo:
```

```python
u=vector([1,1,0]); v=vector([3,0,1]); show(u); show(v)
```

```python
# Sacamos ahora una base ortogonal de P haciendo Gram-Schmidt:
```

```python
u1=u;
```

```python
v1=v-((v*u1)/norm(u1)^2)*u1
```

```python
show(u1); show(v1)
```

```python
# Normalizamos los vectores para tener una base ortonormal:
```

```python
u1=u1.normalized(); v1=v1.normalized(); show(u1); show(v1)
```

```python
# El siguiente vector v2 del plano P forma un ángulo de 60º con u1:
```

```python
a=pi/3 # El ángulo de 60º en radianes.
```

```python
v2=cos(a)*u1+sin(a)*v1; show(v2)
```

```python
# Por tanto, una base de P formada por dos vectores (de norma 1) que forman un ángulo de 60º es {u1,v2}.
```
