# 2d - giro alrededor del origen

```python
# Vamos a considerar el giro de ángulo "a" alrededor del origen, siendo: 
```

```python
a=pi/6 # Ángulo de 30º.
```

```python
# La matriz de dicho giro en base canónica es:
```

```python
M=column_matrix([[cos(a),sin(a)],[-sin(a),cos(a)]]); show(M)
```

```python
# Dibujamos los vectores de la base canónica y sus imágenes por el giro:
```

```python
u=vector([1,0]); v=vector([0,1])
```

```python
plot(u)+plot(M*u,color='green')+plot(v,color='red')+plot(M*v,color='orange')+circle((0,0),1,color='black')
```
